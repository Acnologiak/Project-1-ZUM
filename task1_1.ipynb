# –í–∏–¥–∞–ª–µ–Ω–Ω—è URLs

```python
import re

#re_url = re.compile(r'((www.[^\s]+)|(https?://[^\s]+))')
re_url = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
test_block['Tweet'] = test_block['Tweet'].apply(lambda x: re_url.sub('', x))

test_block.sample(10)
```

# –í–∏–¥–∞–ª–µ–Ω–Ω—è emails

```python
re_email = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b')

test_block['Tweet'] = test_block['Tweet'].apply(lambda x: re_email.sub('', x))
test_block.sample(10)
```

# Lowercase

```python
test_block['Tweet'] = test_block['Tweet'].apply(lambda x: x.lower())
test_block.sample(10)
```

# –í–∏–¥–∞–ª–µ–Ω–Ω—è –ø—É–Ω–∫—Ç—É–∞—Ü—ñ—ó —ñ —Å–ø–µ—Ü—Å–∏–º–≤–æ–ª—ñ–≤

```python
#re_pattern = re.compile(r"[^a-zA-Z\s']") 
re_pattern = re.compile(r"[^a-zA-Z0-9\s]")
test_block['Tweet'] = test_block['Tweet'].apply(lambda x: re_pattern.sub('', x))
test_block.sample(10)
```

```python
import spacy

nlp = spacy.load('en_core_web_md', disable = ['ner', 'parser'])
nlp.add_pipe('sentencizer')

def remove_stopwords(text):
  return ' '.join(filter(lambda x: x not in nlp.Defaults.stop_words, text.split()))

test_block['Tweet'] = test_block['Tweet'].apply(remove_stopwords)
test_block.sample(10)
```

```python
def lemmatize(text):
  return ' '.join([x.lemma_ for x in nlp(text)])

test_block['Tweet'] = test_block['Tweet'].apply(lemmatize)
```
