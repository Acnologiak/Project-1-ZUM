```python
df = pd.read_csv('task_1_2_labeled_data.csv')

df.sample(20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df['Tweet'], df['prediction'], test_size = 0.1, stratify=df['prediction'], random_state=1)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=500000)
vectorizer.fit(X_train)
```

```python
X_train = vectorizer.transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# BernoulliNB

```python
model = BernoulliNB()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(Y_test, y_pred))
```

```python
print(confusion_matrix(Y_test, y_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

decisionTree = DecisionTreeClassifier()

decisionTree.fit(X_train,Y_train)

y_pred = decisionTree.predict(X_test)

print(confusion_matrix(Y_test, y_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

randomForest = RandomForestClassifier(random_state = 42)

randomForest.fit(X_train,Y_train)

y_pred = randomForest.predict(X_test)

print(confusion_matrix(Y_test, y_pred))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logisticRegretion = LogisticRegression()

logisticRegretion.fit(X_train,Y_train)

y_pred = logisticRegretion.predict(X_test)

print(confusion_matrix(Y_test, y_pred))
```
